2. Write a function that inputs a list of integers and returns the minimum value

````python
# Write your function here
def get_min(x):
    current_min = x[0]
    for i in x[1:]:
        if i < current_min:
            current_min = i
    return current_min
````

#### Challenge problem:  
Write a function that take in four arguments: lst1, lst2, str1, str2, and returns a pandas DataFrame that has the first column labeled str1 and the second column labaled str2, that have values lst1 and lst2 scaled to be between 0 and 1.

For example
```
lst1 = [1, 2, 3]
lst2 = [2, 4, 5]
str1 = 'one'
str2 = 'two'

my_function(lst1, lst2, str1, str2)
``` 
should return a DataFrame that looks like:



|  <i></i> | one | two |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | .5 | .666 |
| 2 | 1 | 1 |



````python
import numpy as np
import pandas as pd


def my_function(lst1, lst2, str1, str2):
    
    mx_1 = max(lst1)
    mn_1 = min(lst1)
    lst3 = []
    lst4 = []
    for key in lst1:
        k = ((key-mn_1)/(mx_1-mn_1))
        lst3.append(k)
    mx_2 = max(lst2)
    mn_2 = min(lst2)
    for key in lst2:
        k =((key-mn_2)/(mx_2-mn_2))
        lst4.append(k)
    
    return pd.DataFrame({str1 : lst3 , str2 : lst4})
    
````

````python
lst1 = [1, 2, 3]
lst2 = [2, 4, 5]
str1 = 'one'
str2 = 'two'
my_function(lst1, lst2, str1, str2)
````
